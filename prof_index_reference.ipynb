## データ参照

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
kaisai_date = "20260215"
year = kaisai_date[:4]

RESULT_DIR = Path(".", "prof_result", year)
```

```python
index_results_df = pd.read_csv(RESULT_DIR / f"results_prof_index_{kaisai_date}.csv", encoding="shift-jis")
```

```python
state_list = list(set(index_results_df["場所"].values))
state_list
```

```python
index_results_df[index_results_df["場所"] == state_list[0]] \
    .query("R == 11") \
    .sort_values(by="総合利益度順位")
```
